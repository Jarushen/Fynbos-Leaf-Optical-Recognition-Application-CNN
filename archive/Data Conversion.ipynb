```python
from PIL import Image
import os

def convert_jfif_to_jpg(directory):
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith(".jfif"):
                file_path = os.path.join(root, file)
                image = Image.open(file_path)
                new_file_path = os.path.splitext(file_path)[0] + ".jpg"
                image.save(new_file_path, "JPEG")
                print(f"Converted {file} to JPG")

# Replace 'your_directory_path' with the path to your directory
convert_jfif_to_jpg(r"C:\Users\Jarushen\Desktop\Masters Thesis\Images\Fynbos_Dataset_C")
```

```python
import os

def delete_jfif_files(directory):
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith(".jfif"):
                file_path = os.path.join(root, file)
                os.remove(file_path)  # Delete the JFIF file
                print(f"Deleted {file}")

# Replace 'your_directory_path' with the path to your directory
delete_jfif_files(r"C:\Users\Jarushen\Desktop\Masters Thesis\Images\Fynbos")
```
